**Question 1**: What is Simple Linear Regression (SLR)? Explain its purpose.

Answer:
Simple Linear Regression (SLR) is a statistical method used to model the relationship between a single independent variable (X) and a dependent variable (Y) by fitting a straight line through the data points.
The purpose of SLR is to predict the value of Y based on X and to understand how changes in X affect Y.

**Question 2:** What are the key assumptions of Simple Linear Regression?

Answer:

Linearity: The relationship between X and Y is linear.

Independence: Observations are independent of each other.

Homoscedasticity: The variance of residuals is constant across all values of X.

Normality: Residuals (errors) are normally distributed.

No multicollinearity: (Applies in multiple regression) â€“ variables are not highly correlated.

**Question 6
:**  What is Logistic Regression? How does it differ from Linear Regression?

Answer:
Logistic Regression is used for classification problems where the dependent variable is categorical (e.g., Yes/No, 0/1). It models the probability that Y belongs to a certain category using the logistic (sigmoid) function.

Difference:

Linear Regression: Predicts continuous values.

Logistic Regression: Predicts probabilities or binary outcomes (e.g., 0 or 1).

Linear regression uses a straight line; logistic regression uses an S-shaped (sigmoid) curve.

**Question 7:**  Name and briefly describe three common evaluation metrics for regression models.

Answer:

Mean Absolute Error (MAE): Average of absolute differences between predicted and actual values.

Mean Squared Error (MSE): Average of squared differences between predicted and actual values.

Root Mean Squared Error (RMSE): Square root of MSE; gives error in same units as the dependent variable.

```python
#Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

from sklearn.linear_model import LinearRegression
import numpy as np

# Example data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Create model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
